```python
#Import Packages
# from urllib.request import urlopen as get
# import requests
import time
import pandas as pd
import os
from queue import Queue
#Import Packages

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
# from selenium.webdriver.support.select import Select
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.common.by import By
# from selenium.webdriver.support import expected_conditions as EC


import re
from bs4 import BeautifulSoup
import threading

import sqlite3

# cookies = {}
# def get(url):
#     if not cookies:
#         cookies = requests.head(url)
#     return requests.get(url, cookies=cookies).content
```

```python
conn = sqlite3.connect('Jobs_db_Skills_final.db')
db = conn.cursor()

db.execute('''
          CREATE TABLE IF NOT EXISTS IT_Jobs
          ([job_id] INTEGER PRIMARY KEY NOT NULL, 
          [job_title] TEXT,
          [job_url] TEXT,
          [job_company] TEXT,
          [job_location] TEXT,
          [job_salary] TEXT,
          [job_description] TEXT,
          [job_skills] TEXT)
          ''')


def insert_job(job_title, job_url, job_company, job_location, job_salary, job_description,  job_skills):
    db.execute('''
               INSERT INTO IT_Jobs (job_title, job_url, job_company, job_location, job_salary, job_description,  job_skills)
               VALUES (?, ?, ?, ?, ?, ?, ?)
               ''', (job_title, job_url, job_company, job_location, job_salary, job_description,  job_skills))
    conn.commit()
```

```python

url = 'https://wuzzuf.net/search/jobs/?filters%5Broles%5D%5B0%5D=IT%2FSoftware%20Development'
base_url = 'https://wuzzuf.net'
#Load the web driver and get the url

# driver = webdriver.Chrome(executable_path=r"C:\\Users\\user\\Downloads\\chromedriver_win32\\chromedriver.exe")
```

```python
# driver.get(url)
# page = get(url)

# Specifying incognito mode as you launch your browser[OPTIONAL]
option = webdriver.ChromeOptions()
option.add_argument("--incognito")
option.add_argument("--headless")

# Create new Instance of Chrome in incognito mode
s=Service(ChromeDriverManager().install())
browser = webdriver.Chrome(service=s, options=option)

browser.get(url)
soup = BeautifulSoup(browser.page_source, 'html.parser')

pages_no = soup.find('li', class_ = 'css-8neukt')

# get numbers from string
numbers = re.findall(r'\d+', pages_no.text)
print(numbers)

pages_no =  int( int(numbers[2])/(int(numbers[1]) - int(numbers[0]) + 1) )
pages_no
```

```python
def get_job_info(q,job, job_skills):
    job_link = base_url + job
    browser = webdriver.Chrome(service=s, options=option)
    try:
        browser.get(job_link)
    except:
        print('Error getting the job link')
        return
    soup = BeautifulSoup(browser.page_source, 'html.parser')

    job_title = soup.find('div', class_='css-bjn8wh').find('h1').text
    company_name = soup.find('a', class_ = 'css-p7pghv')
    if company_name:
        company_name = company_name.text
    else:
        company_name= 'Not Specified'

    job_location = soup.find('strong', class_ = 'css-9geu3q')
    if job_location:
        job_location = job_location.text
    else:
        job_location= 'Not Specified'
    
    job_description = soup.find('div', class_ = 'css-1uobp1k')
    if job_description:
        job_description = job_description.text
    else:    
        job_description= 'Not Specified'

    job_skills = soup.find('section', class_='css-3kx5e2').find_all('a', class_='css-g65o95')
    if job_skills:
        job_skills = "".join([skill.text.lower().strip() +'*' for skill in job_skills])
    else:    
        job_skills= 'Not Specified'

    print(job_skills)


    job_salary = soup.find('span', class_ = 'css-4xky9y')
    if job_salary:
        job_salary = job_salary.text
    else:
        job_salary= 'Not Specified'
        
    browser.close()
    q.put([job_title, job_link,company_name, job_location, job_salary, job_description,  job_skills])
    # print([job_title, job_link,company_name, job_location, job_salary, job_description,  job_skills])


shared_q = Queue()
for number in range(pages_no):
    try:
        browser.get(url + '&start=' + str(number))
    except:
        print('Error in getting the url')
        continue
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    job_titles = soup.find_all('div', class_ = 'css-1gatmva e1v1l3u10')


    threads = []
    print(f'start multithreading. Length= {len(job_titles)}, page number= {number}')

    # print(shared_queue.get())
    c = 0
    for job in job_titles:

        job_skills = job.find('div', class_ = 'css-y4udm8').text
        t = threading.Thread(target=get_job_info, args=(shared_q,job.find('a', class_='css-o171kl').get('href'), job_skills))
        t.start()
        threads.append(t)
    for t in threads:
        t.join()
    while (not shared_q.empty()):
        job = shared_q.get()
        insert_job(job[0], job[1], job[2], job[3], job[4], job[5], job[6])
    #     # print(job)
    # break
```
